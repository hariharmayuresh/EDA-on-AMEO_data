# Employement Statistics
Exploratory Data Analysis(EDA)
aspiring_minds_employability_outcomes_2015

# Step 1:
This dataset, obtained from Aspiring Minds' Aspiring Mind Employment Outcome 2015 (AMEO) study, focuses on engineering students. It includes information about the employment outcomes of these graduates, such as their salaries, job titles, and job locations. Additionally, the dataset provides standardized scores in three areas: cognitive skills, technical skills, and personality skills. Demographic features are also included. With approximately 40 independent variables and 4000 data points, the dataset encompasses both continuous and categorical variables. Each candidate is assigned a unique identifier.

Our objective is to conduct exploratory data analysis (EDA) on this dataset and verify certain claims using the available data.

# Step 2:
Import the data and display the head, shape and description of the data.

# Univariate Analysis of Continous Variables: PDF, Histograms, Boxplots.
1) Univariate Analysis on ID Column:

```python
# PDF
pdf = stats.norm.pdf(df['ID'].sort_values(), np.mean(df['ID']), np.std(df['ID']))
plt.plot(df["ID"].sort_values(), pdf, color = 'violet')
plt.xlabel("ID", size=12)    
plt.ylabel("Frequency", size=12)                
plt.grid(True, alpha=0.3, linestyle="--")
plt.show()
```

```python
# HISTOGRAM
sns.histplot(df['ID'])
```

```python
# BOXPLOT
df['ID'].plot.box()
```

Observations: There is not any skeweness and there are no Outliers in the ID column.
